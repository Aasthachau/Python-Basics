##Question 5. What is mutability in Python


What is Mutability in Python?

Mutability in Python refers to whether the value of an object can be changed after it is created.

->If an object can be changed, it is called mutable.

->If an object cannot be changed, it is called immutable.

Mutability is an important concept because it affects how Python stores data and how variables behave inside functions.

1. Mutable Objects (Changeable)-> These objects allow modification of their contents.
You can add, remove, or modify elements.

Examples: list, dictionary, set

2. Immutable Objects (Not Changeable)-> Once created, their values cannot be changed.

Examples: int, float, string, tuple

Why Mutability Matters?

->It affects how variables behave when passed to functions

->It decides whether changes affect the original object

->It improves memory management and performance

## Question 6. Why are lists mutable, but tuples are immutable


Why are lists mutable, but tuples are immutable?

In Python, lists are mutable and tuples are immutable because of the way they are designed and used.

Purpose and Design->
Lists are meant for data that can change. Example: adding, removing, updating items. So Python allows modification → mutable.

Tuples are meant for fixed data collections. Example: coordinates, settings, constants. So Python prevents modification → immutable.

Memory Allocation->
Lists: Use a flexible memory structure that can grow or shrink → so items can be changed.

Tuples: Use a fixed memory structure → once created, nothing can be added, removed, or changed.

Performance->
Tuples are faster and require less memory because they never change.

Lists are slower because they need extra memory to support modifications.

##Question 10. What is the difference between implicit and explicit type casting

Difference Between Implicit and Explicit Type Casting in Python

Python supports two types of type conversion:

1. Implicit Type Casting (Automatic Conversion)->

Python automatically converts one data type to another.

Happens when there is no data loss.

Usually occurs during mathematical operations.

Example:

a = 5       # int


b = 2.5     # float


c = a + b   # int + float -> float


print(c)          # 7.5


print(type(c))    # float


2. Explicit Type Casting (Manual Conversion)->

You manually convert one data type to another using functions.

Use functions like int(), float(), str(), etc.

Needed when Python cannot convert automatically.

Example:

x = "10"


y = int(x)   # converting string to int manually


print(y)   #10


print(type(y))     # int

## Question 13. What is the difference between for and while loops

Difference Between for Loop and while Loop in Python
1. Purpose

for loop

Used when you know the number of iterations.

Best for looping over a sequence (list, string, range, etc.).

while loop

Used when you don’t know how many times the loop will run.

Continues until a condition becomes False.

2. How They Work?

for loop->

Runs for each item in a sequence.

Example:

for i in range(5):

    print(i)

while loop->

Runs as long as the condition is True.

Example:

i = 0


while i < 5:


    print(i)
    i += 1

3. Simple Understanding

for loop: “Repeat this a specific number of times.”

while loop: “Repeat this until something happens.”

```python
# 14. Write a program to reverse a string using a while loop.

text = input("Enter a string: ")

reverse_str = ""
i = len(text) - 1

while i >= 0:
    reverse_str += text[i]
    i -= 1

print("Reversed string:", reverse_str)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num=int(input("Enter a number:"))
fact=1
i=1

while i<=num:
  fact*=i
  i+=1

print(f"factorial of {num} is {fact} ")
```
